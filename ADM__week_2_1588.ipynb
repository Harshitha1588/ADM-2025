```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/_titanic.csv')
```

```python
df
```

```python
plt.hist(df['Age'],bins=30,color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('basic histogram')
```

```python
# Calculate frequency of each age
age_counts = df['Age'].value_counts().sort_index() # Sort by age for a better visualization

# Create bar chart
plt.bar(age_counts.index, age_counts.values, color='black', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('bar chart')
```

```python
# Create box plot
plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True,
            boxprops={'facecolor': 'skyblue', 'edgecolor': 'black'},
            medianprops={'color': 'black'},
            whiskerprops={'color': 'green'},
            capprops={'color': 'red'})
plt.xlabel('Age')
plt.title('Box Plot')
plt.show()
```

```python

sns.pairplot(df)
plt.title('Pair Plot')
plt.show()
```

```python
# Select only the numerical features for correlation analysis
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix for numerical features
corr = numerical_df.corr()

# Generate the heatmap using seaborn
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x=df['Age'], color='skyblue')
plt.xlabel('Age')
plt.title('Violin Plot')
plt.show()
```

```python
sns.countplot(x='Age', data=df, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Count Plot')
plt.show()
```

```python
# Group by Age and count occurrences
age_counts = df['Age'].value_counts().sort_index()

# Create line chart
plt.plot(age_counts.index, age_counts.values, color='skyblue', marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Line Chart')
plt.show()
```
